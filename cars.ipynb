```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'dataset/cars.csv'

# Cargar el archivo CSV en un DataFrame de Pandas
df = pd.read_csv(file_path)

# Imprimir las primeras filas del DataFrame
df.head()
```

```python
# EXPLORACION DEL DATASET

print("Filas/Columnas: ", df.shape)

print("Tipos de Datos: ")
df.dtypes
```

```python
# Eliminamos las columnas 'feature_X'

df = df.drop("feature_0", axis='columns')
df = df.drop("feature_1", axis='columns')
df = df.drop("feature_2", axis='columns')
df = df.drop("feature_3", axis='columns')
df = df.drop("feature_4", axis='columns')
df = df.drop("feature_5", axis='columns')
df = df.drop("feature_6", axis='columns')
df = df.drop("feature_7", axis='columns')
df = df.drop("feature_8", axis='columns')
df = df.drop("feature_9", axis='columns')
df.shape
```

```python
#df.head()
#df.tail()
df.sample(5)
```

# Análisis de Valores Numéricos

```python
# HISTOGRAMAS - Permiten agrupar números en rangos (o contenedores).
# (se usan los argumentos 'figsize' y 'xrot' para aumentar el tamaño de la cuadrícula y girar el eje x 45 grados respectivamente)

df.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
# RESUMEN ESTADÍSTICO DE LOS DATOS NUMÉRICOS

df.describe()
```

# Análisis de Valores Categóricos

```python
# RESUMEN ESTADÍSTICO DE VALORES CATEGÓRICOS

df.describe(include='object')
```

```python
# SE GRAFICA CADA VARIABLE CATEGÓRICA (CONTEO DE CADA CLASE)

for column in df.select_dtypes(include='object'):
    if df[column].nunique() < 10:
        sns.countplot(y=column, data=df)
        plt.show()
```

# Agrupación y Segmentación
### La segmentación nos permite cortar los datos y observar la relación entre características categóricas y numéricas.

```python
# SEGMENTACIÓN DE LA VARIABLE OBJETIVO POR CARACTERÍSTICAS CATEGÓRICAS
# Se compara la característica objetivo ('price_usd'), entre las diversas clases de las principales características categóricas (odometer_value, year_produced, engine_capacity) para ver cómo cambia el precio con las clases.

for column in df.select_dtypes(include='object'):
    if df[column].nunique() < 10:
        sns.boxplot(y=column, x='price_usd', data=df)
        plt.show()
```

```python
# AGRUPACIÓN DE CARACTERÍSTICAS NUMÉRICAS POR CADA CARACTERÍSTICA CATEGÓRICA
# Se muestra cómo todas las demás características numéricas, no sólo el precio, cambian con cada característica categórica al resumir las características numéricas en todas las clases.

for column in df.select_dtypes(include='object'):
    if df[column].nunique() < 10:
        display(df.groupby(column).mean())
```

```python
# CORRELACIONES - Relaciones entre características numéricas.
# Se muestra la matriz de correlaciones para las diferentes características numéricas.

corrs = df.corr()
corrs
```

```python
# Mapa de calor de las correlaciones

plt.figure(figsize=(10,8))
sns.heatmap(corrs, cmap='RdBu_r', annot=True)
plt.show()
```

```python
# TOP 10 de Fabricantes con más vehículos a la venta

top10_fabricantes = (df['manufacturer_name'].value_counts()).iloc[:10]
top10_fabricantes.plot(kind='bar', title="Top 10 fabricantes")
```

```python
# Colores disponiles (Top 5)

top10_colores = (df['color'].value_counts()).iloc[:5]
top10_colores.plot(kind='pie', title="Colores disponibles", autopct='%1.1f%%')
```

```python
# GRÁFICO DE DISPERSIÓN - Precio vs. Año de fabricación

year_prod = df['year_produced']
price = df['price_usd']

plt.figure(figsize=(8, 6))
sns.scatterplot(x=year_prod, y=price)
plt.title('Relación entre año de producción y precio')
plt.xlabel('Año de producción')
plt.ylabel('Precio')
plt.show()
```

```python
# GRÁFICO DE DISPERSIÓN - Precio vs. kilometraje

odometer = df['odometer_value']
price = df['price_usd']

plt.figure(figsize=(8, 6))
sns.scatterplot(x=odometer, y=price)
plt.title('Relación entre kilometraje y precio')
plt.xlabel('Odometro')
plt.ylabel('Precio')
plt.show()
```
